```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from crewai import LLM

llm = LLM(
    model = "gemini/gemini-2.0-flash",
    temperature=0.1
)
```

```python
from crewai import Agent,Task,Crew
from crewai_tools import TavilySearchTool

search_tool = TavilySearchTool()

researcher=Agent(
    role="You are an expert researcher about trending and lastest updates and news about {topic}",
    goal= "Research and gather the latest information, news, and updates about the given topic from reliable sources.",
    backstory="You are a diligent researcher with a keen interest in staying updated about {topic}. You excel at finding accurate and relevant facts from credible sources.",
    verbose = True,
    tools=[search_tool],
    llm=llm
)

writter=Agent(
    role="You are an expert LinkedIn content writer using the given facts and information to create engaging posts.",
    goal="Create engaging, inspiring, and informative LinkedIn posts about trending technologies, events, tips, case studies.",
    backstory="You are a enthusiastic content creator whose is interested in wrtting interesting, gripping, beneficial LinedIn  posts on technology and programming topics.",
    verbose = True,
    llm=llm
)
```

```python
task1 = Task(
    description="Find interesting and trending updates about the given topic from reliable sources.",
    expected_output="A bulleted list of the latest and most interesting updates about the given topic.",
    agent=researcher
)

task2=Task(
    description="Using the researched information, create an engaging LinkedIn post.",
    expected_output="A well-structured LinkedIn post that is informative and engaging.",
    agent=writter,
    context=[task1]
)
```

```python
crew = Crew(
    agents=[researcher, writter],
    tasks=[task1, task2],
    verbose=True
)

result = crew.kickoff(inputs={"topic": "decreasing value of rupees"})  # kickoff may return hex/bytes output

print(result)
```
